```python
# Function to rename duplicate columns
def rename_duplicates( old_cols ):
    seen = set()
    for item in old_cols:
        fudge = 1
        newitem = item
        while newitem in seen:
            fudge += 1
            newitem = "{}_{}".format(item, fudge)
        yield newitem
        seen.add(newitem)

# Apply renaming to your DataFrame columns to ensure uniqueness
AML_subtype_df.columns = list(rename_duplicates(AML_subtype_df.columns))
```

```python
# Re-merge using an explicit approach, ensuring 'ID_REF' from cohort is used for merge without duplication
AML_subtype_df = Final_annotated_RNAseq_df.merge(cohort[['ID_REF','SUBTYPES']].drop_duplicates('ID_REF'), left_index=True, right_on='ID_REF', how='left')

# Apply the function to handle any unintended duplicate columns resulting from the merge
AML_subtype_df.columns = list(rename_duplicates(AML_subtype_df.columns))

#fix for issue with column indexing
columns = ['ID_REF', 'SUBTYPES'] + [col for col in AML_subtype_df.columns if col not in ['ID_REF', 'SUBTYPES']]
AML_subtype_df = AML_subtype_df[columns]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
import numpy as np

features_complete = complete_genes.drop(['ID_REF', 'SUBTYPES'], axis=1)

# Scale your features
scaler = StandardScaler()
features_complete_scaled = scaler.fit_transform(features_complete)

def calculate_dvindex(n_clusters, data_scaled):
    if n_clusters == 1:
        return np.inf  # Can't properly calculate DVIndex for a single cluster
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(data_scaled)
    labels = kmeans.labels_
    cluster_centers = kmeans.cluster_centers_
    
    intra_distances = pairwise_distances(data_scaled, cluster_centers[labels], metric='euclidean')
    intra_compactness = np.mean(np.min(intra_distances, axis=1))
    
    if n_clusters > 1:
        inter_distances = pairwise_distances(cluster_centers, metric='euclidean')
        np.fill_diagonal(inter_distances, np.inf)  # Ignore zero distance to itself
        inter_separation = np.min(inter_distances)
    else:
        inter_separation = np.inf  # Undefined for a single cluster
    
    dv_index = intra_compactness / inter_separation
    return dv_index

# Determine the optimal number of clusters based on DVIndex
dvindex_values = []
max_clusters = 20  # You might adjust this based on your data and needs

for k in range(1, max_clusters + 1):
    dv_index = calculate_dvindex(k, features_complete_scaled)
    dvindex_values.append(dv_index)

optimal_clusters = np.argmin(dvindex_values) + 1  # +1 because of zero-based index offset
print(f"Optimal number of clusters: {optimal_clusters}")

# Final clustering with the optimal number of clusters
kmeans_final = KMeans(n_clusters=optimal_clusters, random_state=42).fit(features_scaled)
complete_genes['Cluster'] = kmeans_final.labels_

```
